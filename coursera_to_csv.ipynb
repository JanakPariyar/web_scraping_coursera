```python
user_input=[
'online_education','social_science','language_training','teacher_training','test_prep','teaching_and_academics','logic'
]
```

```python
# Define global variables for page number and max pages
index=0
course_data=[]
while index < len(user_input):
    print(user_input[index])
    page_number=1
    while page_number<=84:
        user_input[index] = user_input[index].lower().replace(' ', '%20')
        query_url = f"https://www.coursera.org/search?query={user_input[index]}&page={page_number}&sortBy=BEST_MATCH"
        r = requests.get(query_url)
        if r.status_code == 200:
            print(page_number,"okay")
            soup = bs(r.text, "html.parser")
            content_boxes = soup.find_all("div", class_="cds-ProductCard-content")


            for content_box in content_boxes:
                course_info = {'user_input': '', 'course_partner': '', 'course_title': '', 'skill_gain': '', 'course_rating': '', 'course_review': '', 'course_detail': ''}
                course_info['user_input'] = user_input[index]
                content_bodies=content_box.find_all("div",class_="cds-CommonCard-bodyContent")
                course_info['course_partner'] = content_box.find("p", class_="cds-ProductCard-partnerNames css-vac8rf").text.strip()
                course_info['course_title'] = content_box.find("h3", class_="cds-CommonCard-title css-6ecy9b").text.strip()
                content_reviews=content_box.find_all("div",class_="css-1vsx0as")
                course_details=content_box.find_all("div",class_="cds-CommonCard-metadata")

                for course_detail in course_details:
                    course_info['course_detail']=course_detail.find("p",class_="css-vac8rf").text.strip()

                for content_body in content_bodies:
                    course_info['skill_gain'] = content_body.find("p", class_="css-vac8rf").text.strip()

                for content_review in content_reviews:
                    course_content_info = content_review.find("p", class_="css-2xargn")

                    if course_content_info is not None:
                        rating_text = course_content_info['aria-label']
                        course_info['course_rating'] = rating_text.split()[0]
                    else:
                        course_info['course_rating'] = 'NaN'

                    course_content_review = content_review.find("p", class_="css-vac8rf")
                    if course_content_review is not None:
                        course_info['course_review'] = course_content_review.text.strip()
                    else:
                        course_info['course_review'] = 'NaN'

                # Append course_info to the course_data list
                course_data.append(course_info)
            # Write course data to a CSV file
            csv_filename = f'course_data_{user_input[index]}.csv'
            with open(csv_filename, 'a', newline='', encoding='utf-8') as csvfile:
                fieldnames = ['user_input', 'course_title', 'course_partner', 'course_review', 'skill_gain', 'course_rating', 'course_detail']
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writeheader()
                for course_info in course_data:
                    writer.writerow(course_info)
        page_number +=1
    index+=1

#  #taking input of desired course or field

# # Create a Tkinter window
# window = tk.Tk()
# window.title("Input Processor")

# # Create an entry widget for user input
# entry = tk.Entry(window, width=30)
# entry.pack(pady=10)

# # Create a submit button

# submit_btn = tk.Button(window, text="Submit", command=process_input)
# submit_btn.pack()

# # Run the Tkinter event loop
# window.mainloop()
```
